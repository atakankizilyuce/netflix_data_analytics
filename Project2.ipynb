## Finding outliers using statistical methods

### Because the data do not follow a normal distribution, we will calculate outlier data points using a statistical method called interquartile range (IQR) instead of using a Z-score. Using IQR, outlier data points are points that fall below Q11.5 IQR or above Q3 + 1.5 IQR. Q1 is the 25th percentile and Q3 is the 75th percentile of the dataset, and IQR represents the interquartile range calculated from Q3 minus Q1 (Q3Q1).

### Using the handy pandas .quantile() function, we can create a simple Python function that takes our column from the dataframe and returns the outliers:

```python
def find_outliers_IQR_Runtime(df):

   q1_runtime=df.quantile(0.25)

   q3_runtime=df.quantile(0.75)

   IQR_runtime=q3_runtime-q1_runtime

   outliers_runtime = df[((df<(q1_runtime-1.5*IQR_runtime)) | (df>(q3_runtime+1.5*IQR_runtime)))]

   return outliers_runtime

outliers_runtime = find_outliers_IQR_Runtime(df['Runtime'])

print('number of outliers: '+ str(len(outliers_runtime)))

print('max outlier value: '+ str(outliers_runtime.max()))

print('min outlier value: '+ str(outliers_runtime.min()))

outliers_runtime
```

```python
def find_outliers_IQR_IMDB(df):

   q1_imdb=df.quantile(0.25)

   q3_imdb=df.quantile(0.75)

   IQR_imdb=q3_imdb-q1_imdb

   outliers_imdb = df[((df<(q1_imdb-1.5*IQR_imdb)) | (df>(q3_imdb+1.5*IQR_imdb)))]

   return outliers_imdb

outliers_imdb = find_outliers_IQR_IMDB(df['IMDB Score'])

print('number of outliers: '+ str(len(outliers_imdb)))

print('max outlier value: '+ str(outliers_imdb.max()))

print('min outlier value: '+ str(outliers_imdb.min()))

outliers_imdb
```
